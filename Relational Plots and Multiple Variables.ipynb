```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

housing=pd.read_csv("..\\my_datasets\\housing.csv")
print(housing.head(2))
print('-----------------')
print(housing.tail())
```

```python
housing.info()
```

```python

sns.relplot(data=housing , x='Gr Liv Area' , y ='SalePrice')
```

```python
sns.set_theme()

sns.regplot(data=housing , x='Gr Liv Area' , y ='SalePrice')
```

```python
sns.relplot(data=housing , x='Gr Liv Area' , y ='SalePrice' , hue='Overall Qual',palette='RdYlGn')
```

```python
sns.relplot(data=housing , x='Gr Liv Area' , y ='SalePrice' , hue='Overall Qual',palette='RdYlGn', size='Garage Area' , sizes=(1,300))
```

```python
sns.relplot(data=housing , x='Gr Liv Area' , y ='SalePrice' , hue='Overall Qual',palette='RdYlGn', style='Rooms', size='Garage Area' , sizes=(1,300,),markers=['*','v'])
```

```python
sns.relplot(data=housing , x='Gr Liv Area' , y ='SalePrice' , hue='Overall Qual',palette='RdYlGn', style='Rooms', size='Garage Area' , sizes=(1,300,),col='Year')
```
